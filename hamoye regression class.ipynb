```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ["features", col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
lr_weights = get_weights_df(lr, X_train, 'lr_weights')
lr_weights
```

```python
rr_weights = get_weights_df(rr, X_train, 'rr_weights')
rr_weights
```

```python
regressor_weights = get_weights_df(regressor, X_train, 'regressor_weights')
regressor_weights
```

```python
final_weights = pd.merge(lr_weights, rr_weights, on = 'features')
final_weights = pd.merge(final_weights, regressor_weights, on = 'features')
final_weights
```
